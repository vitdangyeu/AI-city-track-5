```python
import os
from ultralytics import YOLO
import matplotlib.pyplot as plt
import pandas as pd
import time 

root_dir = '/kaggle/input/aicity-track5-test'
path_gt_csv = '/kaggle/input/prediction-yolo/prediction_yolo/prediction_yolo/Data/Ground_truth/Test/gt_yolo_test.csv'
model = YOLO('/kaggle/input/prediction-yolo/prediction_yolo/prediction_yolo/Data/Weights/best_5.pt')
df = pd.read_csv(path_gt_csv)
```

```python
prediction = []
for index, row in df.iterrows():
    
    img_path = os.path.join(root_dir,'0' * (3-len(str(row['video_id']))) + f"{row['video_id']}", f"frame{row['frame']}.jpg")
    results = model.predict(
        img_path,
        verbose=False
    )
    result = results[0]
    for box in result.boxes:
        class_id = int(box.cls[0].item()) + 1
        bbox = box.xywh[0]
        bbox[:2] = bbox[:2] - bbox[2:]/2
        cords = bbox.tolist()
        cords = [round(x) for x in cords]
        conf = round(box.conf[0].item(), 2)
        if conf > 0.25:
            prediction.append([row['video_id'], row['frame'] + 1, cords[0], cords[1], cords[2], cords[3], class_id , conf])
    if (index + 1) %100 == 0:
        print(f"Done {index + 1}")
```

```python
# TÃªn tá»‡p Ä‘á»ƒ lÆ°u
file_name = "/kaggle/working/gt_pred_yolo5.txt"

# Ghi danh sÃ¡ch vÃ o tá»‡p vÄƒn báº£n
with open(file_name, 'w') as file:
    for sublist in prediction:
        # Chuyá»ƒn má»—i pháº§n tá»­ trong sublist thÃ nh chuá»—i
        sublist = [str(item) for item in sublist]
        # Ghi sublist vÃ o tá»‡p, phÃ¢n tÃ¡ch cÃ¡c pháº§n tá»­ báº±ng dáº¥u pháº©y
        file.write(','.join(sublist) + '\n')
```
